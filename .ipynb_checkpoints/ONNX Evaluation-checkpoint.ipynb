# ONNX Inferencing
The purpose of this session is to evaluate the ONNX model generated in Training session.

We will use this ONNX to create Tensorrt to run on NVidia GPU.

#### Inferencing using ONNX

##### Import required libraries

```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
import cv2
import onnx
import onnxruntime
```

Show the image used for inferencing. You can replace "rambo.jpg" with "wick.jpg".

```python
img = cv2.imread('wick.jpg')
plt.imshow(img)
plt.show()
```

Resize input image to proper size. We used 256*256 when we trained the model, hence this line resize input image to the similar size.

```python
resize = tf.image.resize(img, (256,256))
plt.imshow(resize.numpy().astype(int))
plt.show()
```

Create an ONNX Inference Session. We use this session to load the ONNX model and run inference on it.

```python
model = "model.onnx"
session = onnxruntime.InferenceSession(model, None)
```

To execute Inferencing on ONNX model, you need to know input name of the model. 

You could find input and output name manually using Netron, however easier path is to use ONNX command like the folloeing Cell.

```python
input_name = session.get_inputs()[0].name
output_name = session.get_outputs()[0].name
print(input_name)
print(output_name)
```

Run the inference and get the result in the following Cell.

```python
result = session.run([output_name], {input_name: np.expand_dims(resize/255, 0)})
print(result[0][0][0])
if result[0][0][0] > 0.5: 
    print(f'Predicted class is Wick')
else:
    print(f'Predicted class is Rambo')
```
